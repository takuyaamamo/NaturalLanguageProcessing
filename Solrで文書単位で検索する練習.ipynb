### 文書単位で検索する

```python
import json
import sys
sys.path.append("src")
import solrindexer as indexer

# 検索を実行
results = indexer.search(keywords=[['アメリカ'], ['大学']], rows=5)

print('responseHeader')
print(json.dumps(results['responseHeader'], indent=4, ensure_ascii=False), '\n\n')

# 入力したキーワードを含む部分を抜き出して、ハイライトしたデータがl入ってる
print('highlighting')
print(json.dumps(results['highlighting'], indent=4, ensure_ascii=False), '\n\n')

print('response')
# nuFoundはヒットしたデータの数を表す。
print(results['response']['numFound'])
# docsには検索結果が入っている。
for row in results['response']['docs']:
    for fl, value in row.items():
        if fl == 'content_txt_ja':
            value = value[:300].replace('\n', '')
        print(f'{fl}\t{value}')
    print()
```
